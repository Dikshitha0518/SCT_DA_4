```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
df=pd.read_csv("marketing_campaign.csv")
```

```python
df.head()
df.shape
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
df["Income"].fillna(df["Income"].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['Age']=2025-df['Year_Birth']
```

```python
df['Response'].value_counts().plot(kind='bar')
plt.title('Campaign Response Distribution')
plt.xlabel('Response')
plt.ylabel('Count')
plt.show()
```

```python
df.groupby('Response')['Income'].mean().plot(kind='bar')
plt.title('Average Income by Campaign Response')
plt.xlabel('Response')
plt.ylabel('Average Income')
plt.show()
```

```python
df.groupby('Marital_Status')['Response'].mean().plot(kind='bar')
plt.title('Response Rate by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Average Response Rate')
plt.show()
```

```python
df.groupby('Education')['Response'].mean().plot(kind='bar')
plt.title('Response Rate by Education')
plt.xlabel('Education')
plt.ylabel('Average Response Rate')
plt.show()
```

```python
df.groupby('Response')['NumWebPurchases'].mean().plot(kind='bar')
plt.title('Average Web Purchases by Response')
plt.xlabel('Response')
plt.ylabel('Web Purchases')
plt.show()
```

```python
df.groupby('Response')['NumStorePurchases'].mean().plot(kind='bar')
plt.title('Average Store Purchases by Response')
plt.xlabel('Response')
plt.ylabel('Store Purchases')
plt.show()
```

```python
funnel={
    'Web Visits':df['NumWebVisitsMonth'].sum(),'WebPurchases':
    df['NumWebPurchases'].sum(),'Store Purchases':
    df['NumStorePurchases'].sum()
}
plt.bar(funnel.keys(),funnel.values())
plt.title('Marketing Funnel')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.select_dtypes(include=np.number).corr(),cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

Business Insights Report (EDA)

Objective

The objective of this analysis is to explore the marketing campaign dataset, identify customer behavior patterns, and recommend which marketing channels should receive more budget based on customer engagement and expected return on investment (ROI).

Data Cleaning

- Checked the dataset for missing values and duplicates.
- Missing values in the Income column were handled.
- Duplicate records were removed.
- A new Age column was created from the Year_Birth column.

Key Findings

1. Customers with higher income showed a higher campaign response rate.
2. Most customers belong to the age group of 40–65 years.
3. Single and divorced customers responded better than married customers.
4. Customers who made more web purchases were more likely to respond to marketing campaigns.
5. The correlation analysis showed strong relationships among product spending categories.

Marketing Funnel Analysis

The analysis indicates that many customers visit the website, but only a smaller percentage complete purchases. Improving conversion from website visits to purchases can increase overall campaign performance.

Recommendations

- Increase budget for digital marketing and web-based campaigns.
- Target high-income customers with personalized offers.
- Focus remarketing campaigns on customers with previous online purchases.
- Optimize website experience to improve conversions.
- Continue monitoring campaign performance and ROI before increasing spending on lower-performing channels.

Conclusion

Exploratory Data Analysis shows that digital channels and high-value customer segments provide the best opportunity for improved marketing ROI. Future marketing budgets should prioritize online campaigns while targeting customers with higher purchasing potential.
